```python
df['detect'] = df['spam'].apply(lambda x:1 if x=='spam' else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.spam,df.detect,test_size = 0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
V = CountVectorizer()
```

```python
x_train_count = V.fit_transform(x_train.values)
```

```python
x_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train_count,y_train)
```

```python
emails = ["Nah I don't think he goes to usf, he lives around here though","Nah I don't think he goes to usf, he lives around here though"]
```

```python
emails_count = V.transform(emails)
```

```python
model.predict(emails_count)
```

```python
x_test_count = V.transform(x_test)
```

```python
model.score(x_test_count,y_test)
```
